Penjelasan tentang indexing pada array 2D

```python
import numpy as np
x=[[4,3,2,1],[1,2,3,4],[5,3,1,0]]
x_ar=np.array(x)

print(x_ar)
```

```python
x_s1=x_ar[0][0:4] #baris 0 semua kolom, kolom 0-3
print(x_s1)
```

```python
x_s2=x_ar[0:2][0:2] #baris 0-1, kolom 0-1
print(x_s2)
```

```python
x_ar[0:2,0:2]=[[-1,-1],[-2,-2]]
print(x_ar)
```

```python
x_s3=x_ar[x_ar>0]
print(x_s3)
```

```python
x_s4=x_ar[:,0:1]
print(x_s4)
```

```python
x_s4=x_ar[:,:] #seluruh baris kalau :,:
print(x_s4)
```

```python
x_s5=x_ar[:2,0:1] 
print(x_s5)
```

```python
x_s5=x_ar[:2,0:1] #:2 artinya baris awal atau s sd (2-1)
print(x_s5)
```

```python
x_s5=x_ar[:,:1] #:2 artinya baris awal atau s sd (2-1)
print(x_s5)
```

```python
x_s5=x_ar[:,:2] #:2 artinya baris awal atau s sd (2-1)
print(x_s5)
```

```python
x=[[4,3,2,1,0,1,2],[1,2,3,4,-1,-1,-1],[5,3,1,0,-2,-2,-2]]
xa=np.array(x)
print(xa)
```

```python
x1=xa[:,:2] #:2 artinya baris awal atau s sd (2-1)
print(x1)
x2=xa[:,:-2] #:-2 artinya kolom yg ditampilkan  tdk termasuk 2 kolom terakhir
print(x2)
```

```python
xt=np.transpose(xa) #ukuran awalnya 3x7 emnjadi 7x3
print(xt)
xt.shape
```

```python
#mengubah dimensi, tapi tidak mengubah jumlah elemenya
xr=xt.reshape(1,21)
print(xr)
xr.shape
```

```python
xr=xt.reshape(21,1)
print(xr)
xr.shape
```

```python
import pandas as pd

file='epl-goalScorer(20-21).csv'
db=pd.read_csv(file) #menampilkan 5 baris teratas dan 5 baris terbawah
#ada 521 row
print (db)
```

```python
db.describe() #hanya berlaku untuk tipe yg numerik (float dan int) saja, untuk strik tidak bisa dihitung
```

```python
 #dapat menampilkan tidak hanya numerik, namun juga string, hanya saja tidak bisa dihitung bagian rata" (mean), standart defiasi (std)
#yg paling sering muncul disana terlihat pada bagian TOP yaitu harry kane
#menghasilkan NaN karena tidak terhitung stringnya disana
db.describe(include = 'all')
```

```python
#menghitung rata"
db.mean()
#adanya warning, karena ada string, shg muncul bgt
```

```python
db.std()
```

```python
db.quantile(0.9)
```

```python
q1=db.quantile(0.25)
q3=db.quantile(0.75)
iqr=q3-q1
print(iqr)
```

```python
#mencari nilai pencilan
pencilan=(db<q1-1.5*iqr) | (db>q3+1.5*iqr)
print(pencilan)

```

```python
#value counts adalah jumlah string yg muncul
db.value_counts()
```

```python
#korelasi/hubungan antara 1 kolom dengan kolom lain
#saat kolom yg sama dihitung, maka nanti hasilnya akan 1 karena dia sama
#makin tinggi nilai korelasi, maka makin terkait
#seperti waktu dan permainan memiliki nilai korelasi yg tinggi karena memiliki hubungan yg cukup tinggi
#makin mendekati 0, makins edikit hubungannya
#jika bernilai -, dia berhubungan tapi berkebalikan

db.corr()
```
